```python
'''Project Goal: Summarizing the exposure in the portfolio
Questions to answer: 
A. Overall Portfolio Exposure like Portfolio Composition, Concentration
B. Equity Exposure Questions like sector exposures, cross asset concentration
C.Fixed Income exposure questions like sector exposures, ratings distribution, duration exposures, top fixed-income holdings, risk mix
'''
```

```python
import pandas as pd
```

```python
Portfolio = pd.read_csv('Sample_file.csv', header = 0)
Portfolio
```

```python
#What is the total portfolio exposure?

Total_Portfolio_Exposure = Portfolio['weight'].sum()
print("The total portfolio exposure in percentage: ",Total_Portfolio_Exposure*100)
```

```python
#What kind of asset classes portfolio has?

Asset_class = Portfolio['asset_class'].unique()
Asset_class
```

```python
# How much is allocated to each asset class?

Asset_class_weightage = Portfolio.groupby('asset_class')['weight'].sum()*100
Asset_class_weightage
```

```python
#How many positions are there in portfolio

Total_positions = Portfolio.shape[0]
print("Total no of positions in the portfolio are: ", Total_positions)
```

```python
#How many positions are there in each asset class?

No_of_positions_per_asset_class = Portfolio['asset_class'].value_counts()
No_of_positions_per_asset_class
```

```python
#What are the Top 5 positions by weight?

Top_5_positions = Portfolio.sort_values(by = 'weight', ascending = False).head(5)
Top_5_positions
```

```python
#Sum of the weightage of top 5 positions

portfolio_concentration = Top_5_positions['weight'].sum()*100
print('The weightage of the top 5 positions in the portfolio is:', round(portfolio_concentration,2))
```

```python
#What are the largest issuer-level exposures (combining stock + bond of same issuer)?

split_ticker = Portfolio['ticker'].str.split("_").str[0]
Portfolio['base ticker'] = split_ticker
issuer_concentration = Portfolio.groupby('base ticker')['weight'].sum()*100
issuer_concentration
```

```python
# What is the equity exposure by sector in %

Equity_portfolio = Portfolio[Portfolio['asset_class'] == 'Equity']
Equity_portfolio.groupby('sector')['weight'].sum()*100
```

```python
# Which is the largest equity sector overweight?

Equity_portfolio.groupby('sector')['weight'].sum().sort_values(ascending = False).head(1)*100
```

```python
# Which sectors are underrepresented?

Portfolio.groupby('sector')['weight'].sum().sort_values()*100
```

```python
# Which are the largest equity positions by weight?
Equity_portfolio.sort_values(by='weight', ascending = False)
```

```python
# Which is the largest equity positions by weight?
Equity_portfolio.nlargest(n=1, columns = 'weight')
```

```python
# Are there any sectors with only one name (higher idiosyncratic risk)?

Portfolio['sector'].unique()
```

```python
# Are there issuers where the PM holds both equity and credit?
multi_issuer = Portfolio['base ticker'].value_counts()
multi_issuer = multi_issuer[multi_issuer > 1]
multi_issuer
```

```python
# Combined equity+bond exposure per issuer
Portfolio.groupby('base ticker')['weight'].sum()*100
```

```python
fixed_income_portfolio = Portfolio[Portfolio['asset_class'] == 'Fixed Income'].copy()
fixed_income_portfolio
```

```python
# What % of fixed income exposure is in each sector

exposure_by_sector = fixed_income_portfolio.groupby('sector')['weight'].sum()*100
exposure_by_sector
```

```python
# What % of exposure is in each rating
exposure_by_rating = fixed_income_portfolio.groupby('rating')['weight'].sum()*100
exposure_by_rating
```

```python
# What % of exposure is in investment grade vs non-investment grade

def classify_rating(r):
    Investment_Grade = ['AAA', 'AA', 'A', 'A-']
    if r in Investment_Grade:
        return 'IG'
    else:
        return 'Non-IG'

fixed_income_portfolio['rating category'] = fixed_income_portfolio['rating'].apply(classify_rating)
fixed_income_portfolio.groupby('rating category')['weight'].sum()*100
```

```python
# What % of exposure is in each duration bucket

fixed_income_portfolio.groupby('duration_bucket')['weight'].sum()*100
```

```python
# What % of exposure is in short-term vs long-term bonds

def bond_duration(dur):
    short_term = ['1-3Y', '3-5Y']
    if dur in short_term:
        return 'Short Duration'
    else:
        return 'Long Duration'
fixed_income_portfolio['Bond Duration'] = fixed_income_portfolio['duration_bucket'].apply(bond_duration)
fixed_income_portfolio.groupby('Bond Duration')['weight'].sum()*100
```

```python
# How sensitive is my portfolio towards the interest rates

def avg_duration(dur):
    duration = {'1-3Y': 2, '3-5Y': 4, '5-7Y': 6, '7-10Y': 8.5}
    if dur in duration:
        return duration[dur]
    else:
        return 0

fixed_income_portfolio['avg_duration'] = fixed_income_portfolio['duration_bucket'].apply(avg_duration)
print('The average duration of the portfolio is: ', fixed_income_portfolio['avg_duration'].mean())
```

```python
fixed_income_portfolio
```

```python
#What is the weighted average duration of the portfolio
print('The weighted average duration of the portfolio is: ', (fixed_income_portfolio['weight']*fixed_income_portfolio['avg_duration']).sum()/fixed_income_portfolio['weight'].sum())
```

```python
# Which issuers dominate the credit exposure

fixed_income_portfolio.groupby('base ticker')['weight'].sum().sort_values(ascending = False).head(1)*100
```

```python
# How much exposure is in high yield credit [BB, BBB]

high_yield_exposure = fixed_income_portfolio[fixed_income_portfolio['rating category'] == 'Non-IG']['weight'].sum()*100
print('The high yield credit exposure in the portfolio is: ', high_yield_exposure)
```
